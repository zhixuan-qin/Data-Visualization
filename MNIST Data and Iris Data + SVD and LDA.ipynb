# MNIST Data with SVD Visualization

###### SVD projection visualization on sports groups:

```python
import scipy
import numpy as np
from scipy.io import loadmat
from scipy import sparse
data=loadmat('news20.mat')
labels = data['labels']
newsgroups = data['newsgroups'][0]
keywords=data['keywords']
A=data['A']
nonzero_indices=scipy.sparse.csr_matrix.nonzero(A)
A[nonzero_indices]=1 #convert frequency count into binary form
A_colsum=np.sum(A,axis=0)
#keeping only the words with frequencies between 2 and 939
col_indices = np.where((2<=A_colsum) & (A_colsum <939))
A_trimmed = A[:, col_indices[1]]
keywords_trimmed=keywords[col_indices[1]]
labels_trimmed=labels[A_trimmed.getnnz(1)>0] #remove empty rows
A_trimmed = A_trimmed[A_trimmed.getnnz(1)>0] #remove empty rows
A_colsum_trimmed=A_colsum[col_indices]
n=A_trimmed.shape[0] #total number of documents
weights = np.log(np.divide(n, A_colsum_trimmed)) #calculate the weights
squared_weights = np.power(weights.A,2)
nonzero_inds = np.nonzero(A_trimmed)
A_trimmed[nonzero_inds] = np.take(squared_weights, nonzero_inds[1])
```

```python
from sklearn.preprocessing import normalize
A_trimmed_normalized=normalize(A_trimmed, axis=1, norm='l2')
sport_index=[]
for i in range(len(labels_trimmed)):
    if newsgroups[labels_trimmed[i]-1] in ['rec.autos', 'rec.motorcycles', 
                                   'rec.sport.baseball', 'rec.sport.hockey']:
        sport_index.append(i)
A_sports=A_trimmed_normalized[sport_index,:]
sport_label=labels[sport_index,:]
#conduct SVD on A_sports matrix
u, s, vh = scipy.sparse.linalg.svds(A_sports, k=3, which='LM')
v=vh.transpose()
v2=v[:,[2,0,1]]
A_sports_projected=A_sports.dot(v2)
auto_index=[]
motor_index=[]
baseball_index=[]
hockey_index=[]
for i in range(sport_label.shape[0]):
    if newsgroups[sport_label[i]-1]=='rec.autos':
        auto_index.append(i)
    if newsgroups[sport_label[i]-1]=='rec.motorcycles':
        motor_index.append(i)
    if newsgroups[sport_label[i]-1]=='rec.sport.baseball':
        baseball_index.append(i)
    if newsgroups[sport_label[i]-1]=='rec.sport.hockey':
        hockey_index.append(i)
auto_projected=A_sports_projected[auto_index,:]
motor_projected=A_sports_projected[motor_index,:]
baseball_projected=A_sports_projected[baseball_index,:]
hockey_projected=A_sports_projected[hockey_index,:]
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(auto_projected[:,0], auto_projected[:,1], auto_projected[:,2], 
           color = 'b', marker="*", label='rec.autos')
ax.scatter(motor_projected[:,0],motor_projected[:,1],motor_projected[:,2],
          color='r',marker="^", label='rec.motorcycles')
ax.scatter(baseball_projected[:,0],baseball_projected[:,1],baseball_projected[:,2],
          color='g',marker="o", label="rec.sport.baseball")
ax.scatter(hockey_projected[:,0],hockey_projected[:,1],hockey_projected[:,2],
          color='m',marker="p", label="rec.sport.hockey")
plt.legend(loc='upper left', shadow=True, fontsize='medium', bbox_to_anchor=(1, 0, 0.5, 1))
plt.title('3-Dimension SVD Projection of 4 Sport Groups',fontsize=16)
ax.view_init(30, 0)
plt.show()
```

###### For each of the 4 categories, identify and display the 20 most correlated words with the dominant direction in a bar graph:

```python
import heapq
auto_index2=[]
motor_index2=[]
baseball_index2=[]
hockey_index2=[]
for i in range(labels_trimmed.shape[0]):
    if newsgroups[labels_trimmed[i]-1]=='rec.autos':
        auto_index2.append(i)
    if newsgroups[labels_trimmed[i]-1]=='rec.motorcycles':
        motor_index2.append(i)
    if newsgroups[labels_trimmed[i]-1]=='rec.sport.baseball':
        baseball_index2.append(i)
    if newsgroups[labels_trimmed[i]-1]=='rec.sport.hockey':
        hockey_index2.append(i)
auto2=A_trimmed_normalized[auto_index2,:]
motor2=A_trimmed_normalized[motor_index2,:]
baseball2=A_trimmed_normalized[baseball_index2,:]
hockey2=A_trimmed_normalized[hockey_index2,:]
#conduct 4 svd on the 4 groups
u_a, s_a, vh_a = scipy.sparse.linalg.svds(auto2, k=1, which='LM')
u_m, s_m, vh_m = scipy.sparse.linalg.svds(motor2, k=1, which='LM')
u_b, s_b, vh_b = scipy.sparse.linalg.svds(baseball2, k=1, which='LM')
u_h, s_h, vh_h = scipy.sparse.linalg.svds(hockey2, k=1, which='LM')
vh_a_abs, vh_m_abs,vh_b_abs, vh_h_abs=abs(vh_a),abs(vh_m), abs(vh_b), abs(vh_h)
#extract top 20 vi
top20_vha_index=heapq.nlargest(20, range(len(vh_a_abs[0])), vh_a_abs[0].take)
top20_vhm_index=heapq.nlargest(20, range(len(vh_m_abs[0])), vh_m_abs[0].take)
top20_vhb_index=heapq.nlargest(20, range(len(vh_b_abs[0])), vh_b_abs[0].take)
top20_vhh_index=heapq.nlargest(20, range(len(vh_h_abs[0])), vh_h_abs[0].take)
```

```python
plt.figure(figsize=(10,10))
bar_width = 1
plt.subplots_adjust(hspace=0.2, wspace=0.4)
subplot_positions = [221, 222, 223, 224]
plt.subplot(221)
x = np.arange(20)
y1 = vh_a_abs[0][top20_vha_index]
bars = plt.barh(x, y1)
plt.yticks(x, (k[0][0] for k in keywords_trimmed[top20_vha_index]))
plt.gca().invert_yaxis()
plt.title('rec.autos',fontsize=16)
plt.subplot(222)
y2 = vh_m_abs[0][top20_vhm_index]
bars = plt.barh(x, y2)
plt.yticks(x, (k[0][0] for k in keywords_trimmed[top20_vhm_index]))
plt.gca().invert_yaxis()
plt.title('rec.motorcycles',fontsize=16)
plt.subplot(223)
y3 = vh_b_abs[0][top20_vhb_index]
bars = plt.barh(x, y3)
plt.yticks(x, (k[0][0] for k in keywords_trimmed[top20_vhb_index]))
plt.gca().invert_yaxis()
plt.title('rec.sport.baseball',fontsize=16)
plt.subplot(224)
y4 = vh_h_abs[0][top20_vhh_index]
bars = plt.barh(x, y4)
plt.yticks(x, (k[0][0] for k in keywords_trimmed[top20_vhh_index]))
plt.gca().invert_yaxis()
plt.title('rec.sport.hockey',fontsize=16)
plt.show()
```

# Iris Data with LDA Visualization

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
df = pd.read_csv('./iris.data', sep=",", header = None)
X = df[[0,1,2,3]].values #extract feature information
y = df[[4]].values.ravel() #extract labels
enc = LabelEncoder()
y_encoder = enc.fit(y)
y = y_encoder.transform(y) + 1
#create the mean vector
mean = []
for i in range(1,4):
    mean.append(np.mean(X[y==i], axis=0))
#create within-class scatter matrix
WScatter = np.zeros((4,4))
for i,k in zip(range(1,4), mean):
    class_scatter = np.zeros((4,4)) # scatter matrix for every class
    for row in X[y == i]:
        row, k = row.reshape(4,1), k.reshape(4,1) # make column vectors
        class_scatter += (row-k).dot((row-k).T)
    WScatter += class_scatter  
#create between-class scatter matrix
overall_mean = np.mean(X, axis=0)
BScatter = np.zeros((4,4))
for i,mv in enumerate(mean):  
    n = X[y==i+1,:].shape[0]
    mv = mv.reshape(4,1)
    overall_mean = overall_mean.reshape(4,1)
    BScatter += n * (mv - overall_mean).dot((mv - overall_mean).T)
eig_value, eig_vector = np.linalg.eig(np.linalg.inv(WScatter).dot(BScatter))
top2_eig_vector=eig_vector[:,0:2]
X_projected=np.matmul(X,top2_eig_vector)
```

```python
from matplotlib import pyplot as plt
setosa_index=[]
versicolour_index=[]
virginica_index=[]
for i in range(len(df)):
    if df.iloc[:,4][i]=="Iris-setosa":
        setosa_index.append(i)
    if df.iloc[:,4][i]=="Iris-versicolor":
        versicolour_index.append(i)
    if df.iloc[:,4][i]=="Iris-virginica":
        virginica_index.append(i)
plt.scatter(X_projected[setosa_index,0], X_projected[setosa_index,1], 
           color = 'b', marker="*", label='Setosa')
plt.scatter(X_projected[versicolour_index,0], X_projected[versicolour_index,1], 
           color = 'r', marker="^", label='Versicolour')
plt.scatter(X_projected[virginica_index,0], X_projected[virginica_index,1], 
           color = 'm', marker="o", label='Virginica')
plt.legend(loc='upper left', shadow=True, fontsize='medium', bbox_to_anchor=(1, 0, 0.5, 1))
plt.xlabel('Dimension 1',fontsize="large")
plt.ylabel('Dimension 2',fontsize="large")
plt.title('Iris Data Projection - 2 Dimensional LDA',fontsize=16)
plt.show()
```

# MNIST Data with LDA

###### LDA on digit 4 and 9:

```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
data=loadmat('mnist.mat')
A=data['A']#dimension of A is 70000*784
digit=data['y']#length of digit is 70000
Four_indices = np.argwhere(digit==4)
Fours = A[[r for r, c in Four_indices],:]
Nine_indices=np.argwhere(digit==9)
Nines=A[[r for r,c in Nine_indices],:]
FourNine=np.vstack((Fours,Nines))
averaged_FourNine = np.mean(FourNine, axis=0)
centered_FourNine=FourNine-averaged_FourNine
pca = PCA(.95) #determine the number of k needed to preserve 95% of the total variance
lower_dimensional_data = pca.fit_transform(centered_FourNine)
pca = PCA(n_components=pca.n_components_)
reduced_FourNine = pca.fit_transform(centered_FourNine)
# LDA
reduced_four=reduced_FourNine[0:6824,:]
reduced_nine=reduced_FourNine[6824:13782,:]
mean_reduced_four=np.mean(reduced_four, axis=0)
mean_reduced_nine=np.mean(reduced_nine, axis=0)
centered_reduced_four=reduced_four-mean_reduced_four
centered_reduced_nine=reduced_nine-mean_reduced_nine
S_four=np.matmul(centered_reduced_four.T,centered_reduced_four)
S_nine=np.matmul(centered_reduced_nine.T,centered_reduced_nine)
S_w=S_four+S_nine
v=np.linalg.inv(S_w).dot(mean_reduced_four-mean_reduced_nine)
normed_v = v / np.linalg.norm(v)
projected_four=np.matmul(reduced_four,normed_v)
projected_nine=np.matmul(reduced_nine,normed_v)
plt.scatter(list(range(6824)), projected_four, color = 'b', marker="*",s=3, label='4')
plt.scatter(list(range(6958)), projected_nine, color = 'r', marker="^",s=3, label='9')
plt.legend(loc='upper left', shadow=True, fontsize='medium', bbox_to_anchor=(1, 0, 0.5, 1))
plt.xlabel('Index',fontsize="large")
plt.ylabel('Dimension 1',fontsize="large")
plt.title('PCA (95%)+LDA Projection of Number 4 and 9',fontsize=16)
plt.show()
```

###### LDA on digit 3, 5 and 8:

```python
three_indices=np.argwhere(digit==3)
threes=A[[r for r, c in three_indices],:]
three_label=digit[three_indices[:,0]]
five_indices=np.argwhere(digit==5)
fives=A[[r for r,c in five_indices],:]
five_label=digit[five_indices[:,0]]
eight_indices=np.argwhere(digit==8)
eights=A[[r for r,c in eight_indices],:]
eight_label=digit[eight_indices[:,0]]
ThreeFiveEight=np.vstack((np.vstack((threes,fives)),eights))
ThreeFiveEight_label=np.vstack((np.vstack((three_label,five_label)),eight_label)).ravel()
averaged_ThreeFiveEight = np.mean(ThreeFiveEight, axis=0)
centered_ThreeFiveEight=ThreeFiveEight-averaged_ThreeFiveEight
pca = PCA(.95) #determine the number of k needed to preserve 95% of the total variance
lower_dimensional_data = pca.fit_transform(centered_ThreeFiveEight)
pca = PCA(n_components=pca.n_components_)
reduced_ThreeFiveEight = pca.fit_transform(centered_ThreeFiveEight)
#create the mean vector
mean = []
for i in [3,5,8]:
    mean.append(np.mean(reduced_ThreeFiveEight[ThreeFiveEight_label==i], axis=0))
#create within-class scatter matrix
WScatter = np.zeros((146,146))
for i,k in zip([3,5,8], mean):
    class_scatter = np.zeros((146,146)) # scatter matrix for every class
    for row in reduced_ThreeFiveEight[ThreeFiveEight_label == i]:
        row, k = row.reshape(146,1), k.reshape(146,1) # make column vectors
        class_scatter += (row-k).dot((row-k).T)
    WScatter += class_scatter
#create between-class scatter matrix
overall_mean = np.mean(reduced_ThreeFiveEight, axis=0)
BScatter = np.zeros((146,146))
for i,mv in zip([3,5,8], mean):  
    n = reduced_ThreeFiveEight[ThreeFiveEight_label==i,:].shape[0]
    mv = mv.reshape(146,1)
    overall_mean = overall_mean.reshape(146,1)
    BScatter += n * (mv - overall_mean).dot((mv - overall_mean).T)
eig_value, eig_vector = np.linalg.eig(np.linalg.inv(WScatter).dot(BScatter))
top2_v=eig_vector[:,[0,1]]
ThreeFiveEight_projected=np.matmul(reduced_ThreeFiveEight,top2_v)
plt.scatter(ThreeFiveEight_projected[0: 7141,0], ThreeFiveEight_projected[0: 7141,1], 
           color = 'b', marker="*",s=3, label='3')
plt.scatter(ThreeFiveEight_projected[6824:13782,0], ThreeFiveEight_projected[6824:13782,1], 
           color = 'r', marker="^",s=3, label='5')
plt.scatter(ThreeFiveEight_projected[13454:20279,0], ThreeFiveEight_projected[13454:20279,1], 
           color = 'm', marker="o",s=3, label='8')
plt.legend(loc='upper left', shadow=True, fontsize='medium', bbox_to_anchor=(1, 0, 0.5, 1))
plt.xlabel('Dimension 1',fontsize="large")
plt.ylabel('Dimension 2',fontsize="large")
plt.title('PCA (95%) +LDA Projection of Number 3, 5, and 8',fontsize=16)
plt.show()
```

Comment: compared with PCA, LDA has much better separation between classes.
