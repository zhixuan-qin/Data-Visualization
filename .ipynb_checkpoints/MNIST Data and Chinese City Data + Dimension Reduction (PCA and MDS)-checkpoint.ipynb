# MNIST Data and PCA

##### Averaged digit 5's image:

```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
data=loadmat('mnist.mat')
A=data['A']#dimension of A is 70000*784
digit=data['y']#length of digit is 70000
five_indices = np.argwhere(digit==5)
fives = A[[r for r, c in five_indices],:]
averaged_fives = np.mean(fives, axis=0)
plt.title("Averaged digit 5's image",fontsize=16)
plt.imshow(averaged_fives.reshape((28,28)), cmap='gray')
plt.axis('off')
plt.show()
```

##### PCA and explained variance:

```python
from sklearn.decomposition import PCA
#created centered data matrix of digit 5
centered_five_matrix=fives-averaged_fives
#conduct SVD
u, s, vh = np.linalg.svd(centered_five_matrix, full_matrices=True)
#extract the top 50 singular values
top_50_s=s[0:50]
#calculate variance
pca = PCA()
pca.fit(centered_five_matrix)
tot = sum(pca.explained_variance_)
var_exp = [(i/tot)*100 for i in sorted(pca.explained_variance_, reverse=True)] 
cum_var_exp = np.cumsum(var_exp)
#plotting
fig, (ax1, ax2) = plt.subplots(2, figsize = [8,8])
plt.subplots_adjust(hspace=0.6, wspace=0.3)
ax1.plot(range(1,51), top_50_s, '.g')
ax1.set_title("The first 50 singular values of matrix of digit 5", fontsize=16)
ax1.set(xlabel='No. of singular value', ylabel='Singular values')
ax2.plot(range(1, 785), cum_var_exp, '.b')
ax2.set_title("Explained variance",fontsize=16)
ax2.set(xlabel="No. of singular value", ylabel="percentage")
plt.show()
```

```python
pca = PCA(.9)
lower_dimensional_data = pca.fit_transform(centered_five_matrix)
print(pca.n_components_)
```

Comments: we can see from the above figure that we need at least 75 k to preserve 90% of the total variance of the digit class of 5.

##### PCA component visualization of digit 5:

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(centered_five_matrix)
plt.scatter(principalComponents[:,0], principalComponents[:,1],s=6)
plt.xlabel("Component 1", fontsize='large')
plt.ylabel("Component 2",fontsize='large')
plt.title('Two components PCA of digit 5',fontsize=16)
plt.show()
```

###### Top 20 principals as a image:

```python
top20_direction=vh[0:20]
averaged_top20_direction = np.mean(top20_direction, axis=0)
plt.imshow(averaged_top20_direction.reshape((28,28)), cmap='gray')
plt.axis('off')
plt.show()
```

###### Approximation of digit 5 (1-30k):

```python
fig,ax = plt.subplots(5,6, figsize = [10,5])
fig.suptitle("Approximation of digit 5 (1-30 k)", fontsize=16)
plt.subplots_adjust(hspace=0.1, wspace=0.002)
for i in range(30):
    j, k = i//6, i%6
    vkh=vh[0:i,:]
    vk=vkh.transpose()
    Y=np.matmul(centered_five_matrix,vk)
    Yvkh=np.matmul(Y,vkh)
    approximation=averaged_fives+Yvkh
    ax[j,k].imshow(approximation[0].reshape(28, 28),cmap='gray')
    ax[j,k].axis('off')
plt.show()   
```

###### Top 20 contibutors to v1 as images:

```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
import heapq
data=loadmat('mnist.mat')
A=data['A']#dimension of A is 70000*784
digit=data['y']#length of digit is 70000
two_indices = np.argwhere(digit==2)
twos = A[[r for r, c in two_indices],:]#matrix with all the 2s
averaged_twos = np.mean(twos, axis=0)
centered_two_matrix=twos-averaged_fives#centered 2 matrix
#conduct SVD
u, s, vh = np.linalg.svd(centered_two_matrix, full_matrices=True)
u1=u[:,0]
sigma1=s[0]
v1_coefficient=u1/sigma1
#extract the index of the top 20 contributors
top20_index=heapq.nlargest(20, range(len(v1_coefficient)), v1_coefficient.take)
top20_twos=centered_two_matrix[top20_index,:]
#plotting
fig,ax = plt.subplots(5,4, figsize = [6,6])
fig.suptitle('Top 20 contributors to v1', fontsize=16)
plt.subplots_adjust(hspace=0.1, wspace=0.002)
for i in range(20):
    j, k = i//4, i%4
    ax[j,k].imshow(top20_twos[i].reshape((28,28)), cmap='gray')
    ax[j,k].axis('off')
plt.show()
```

Comments:The top 20 images that contribute the most to v1 looks very similar and share a lot of similar characteristics interms of writing. They also occupy somewhat similar spaces in the image.

###### Bottom 20 contibutors to v1 as images:

```python
#extract the index of the bottom 20 contributors
bottom20_index=heapq.nsmallest(20, range(len(v1_coefficient)), v1_coefficient.take)
bottom20_twos=centered_two_matrix[bottom20_index,:]
#plotting
fig,ax = plt.subplots(5,4, figsize = [6,6])
fig.suptitle('Bottom 20 contributors to v1', fontsize=16)
plt.subplots_adjust(hspace=0.1, wspace=0.002)
for i in range(20):
    j, k = i//4, i%4
    ax[j,k].imshow(bottom20_twos[i].reshape((28,28)), cmap='gray')
    ax[j,k].axis('off')
plt.show()
```

Comments:the images that contribute the least to v1 are very different from each other. Some of them are barely even recognizable as 2. They also occupy different spaces in the image.

###### PCA on digit 0 and 1:

```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
data=loadmat('mnist.mat')
A=data['A']#dimension of A is 70000*784
digit=data['y']#length of digit is 70000
zero_indices = np.argwhere(digit==10)
zeros = A[[r for r, c in zero_indices],:]
one_indices=np.argwhere(digit==1)
ones=A[[r for r,c in one_indices],:]
ZeroOne=np.vstack((zeros,ones))
averaged_ZeroOne = np.mean(ZeroOne, axis=0)
centered_ZeroOne=ZeroOne-averaged_ZeroOne
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(centered_ZeroOne)
principalComponents_zero=principalComponents[0:6903,:]
principalComponents_one=principalComponents[6903:14780,:]
plt.scatter(principalComponents_zero[:,0], principalComponents_zero[:,1], 
            color = 'b', marker="o", s=6, label='Digit 0')
plt.scatter(principalComponents_one[:,0], principalComponents_one[:,1], 
            color = 'r', marker="^",s=6, label='Digit 1')
plt.legend(loc='upper left', shadow=True, fontsize='medium', bbox_to_anchor=(1, 0, 0.5, 1))
plt.xlabel("Component 1", fontsize='large')
plt.ylabel("Component 2",fontsize='large')
plt.title('Two components PCA of digit 0 and 1',fontsize=16)
plt.show()
```

# Cinese City Data and MDS

```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
data=loadmat('ChineseCityData.mat')
city=data['Cities']
distance=data['dists']#matrix with lij
#create matrix L with lij^2
L=np.zeros([len(distance),len(distance)])
for i in range(len(distance)):
    for j in range(len(distance)):
        L[i][j]=distance[i][j]**2
rowsum=np.sum(L, axis=1)
colsum=np.sum(L, axis=0)
overallsum=np.sum(L)
n=len(L)
G=np.zeros([n, n])
for i in range(n):
    for j in range(n):
        G[i][j]=1/2*(rowsum[i]/n+colsum[j]/n-overallsum/(n*n)-L[i][j])
w,v=np.linalg.eig(G)#calculate eigenvalue (w) and eigenvectors (v) of matrix G
top2_ind = np.argsort(w)[::-1][:2]
top2w = w[top2_ind]
top2w_matrix=np.diag(top2w)
sigma_sqrt=np.sqrt(top2w_matrix)
top2v=v[:,top2_ind]#the corresponding top 2 eigenvectors
Y=np.matmul(top2v,sigma_sqrt)
plt.figure(figsize=(10,8))
plt.scatter(Y[:,0], Y[:,1])
for i, label in enumerate(city[0]):
    plt.annotate(label[0],(Y[i][0], Y[i][1]))
plt.title("Multidimensional Scaling of 12 Chinese Cities into Two-dimensional Map",fontsize=16)
plt.xlabel("Euclidean space $y_{i}$",fontsize="large")
plt.ylabel("Euclidean space $y_{j}$",fontsize="large")
plt.show()
```

Comments: the relative positions and distances among the 12 cities are well-mapped compared with the Google map of China.

```python
#calculate the stress number
stress_numerator=0
stress_denominator=0
for i in range(len(distance)):
    for j in range(len(distance)):
        stress_numerator+=(distance[i][j]-np.linalg.norm(Y[i]-Y[j]))**2
        stress_denominator+=(distance[i][j])**2
Stress=np.sqrt(stress_numerator/stress_denominator)
print(Stress)
```

Comments: we can see from the above output that the Kruskal stress value is approximately 0.0805, which is smaller than 0.1. Thus, our model is a good fit, and nicely preserves the pairwise distances.
